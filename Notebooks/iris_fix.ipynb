```python
import pandas as pd
```

```python
csv_url = 'https://www.dropbox.com/scl/fi/s8xeh94lwsz3wc4f21qo6/iris2.csv?rlkey=g6xb8yx626qmjoad1ghb5rgh9&st=005ksws0&dl=1'

iris_data = pd.read_csv(csv_url)
```

```python
display(iris_data)
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression


iris_data = iris_data[iris_data['species_id'].between(1,2)]
print(iris_data.shape)

X = iris_data.drop(columns=['species','species_id'])
y = iris_data['species_id'] - 1

# Linear classifier
model = LogisticRegression()
model.fit(X, y)

coefs = list(model.coef_[0].round(4))
# Convert coefs to python float
coefs = [float(c) for c in coefs]


print(dict(zip(X.columns, coefs)))



```

```python
coefs_dict = {'sepal_length': -0.0579, 'sepal_width': -0.8953, 'petal_length': 2.0453, 'petal_width': 0.8142}

# coefs_df = pd.DataFrame(coefs_dict.items(), columns=['feature', 'coef'])

# Create a dataframe from the dictionary, with the keys as columns and the values as rows
coefs_df = pd.DataFrame(coefs_dict.items(), columns=['feature', 'coef'])

coefs_series = coefs_df[['coef']].values
coefs_series = coefs_df.set_index('feature')['coef']

display(coefs_series)
X @ coefs_series
# coefs_series = coefs_series.reindex(X.columns)
```

```python


display(coefs_series)
# display(coefs_series.info())

multiplication_result = X @ coefs_series

display(multiplication_result)

prediction = (multiplication_result > 0).astype(int) + 1
prediction = prediction.values.flatten()


print(prediction.shape)
print(iris_data['species_id'].shape)

accuracy = (prediction == iris_data['species_id']).mean()
print(accuracy)
```
